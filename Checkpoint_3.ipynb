# Selanjutnya mengecek **outliers** jika ada data yang nilainya jauh lebih rendah ata lebih tinggi

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

# Dikarenakan Presentase nya tidak terlalu tinggi namun variabel penting maka akan dilakukan **imputasi winsoirizing** agar tidak mengubah struktur datanya

```python
columns_to_impute = ["Year", "NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

# Pengecekan ulang

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

# 1. Mengidentifikasi tren penjualan video game berdasarkan platform, genre, dan tahun rilis.


```python
yearly_sales = df.groupby("Year")[["Global_Sales"]].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_sales, x="Year", y="Global_Sales", marker="o", color="b")
plt.title("Tren Penjualan Video Game Global Berdasarkan Tahun Rilis", fontsize=14)
plt.xlabel("Tahun Rilis", fontsize=12)
plt.ylabel("Total Penjualan (Juta Unit)", fontsize=12)
plt.grid(True)
plt.show()
```

```python
platform_sales = df.groupby(["Year", "Platform"])[["Global_Sales"]].sum().reset_index()

plt.figure(figsize=(14, 7))
sns.lineplot(data=platform_sales, x="Year", y="Global_Sales", hue="Platform", linewidth=1.5)
plt.title("Tren Penjualan Video Game Berdasarkan Platform", fontsize=14)
plt.xlabel("Tahun Rilis", fontsize=12)
plt.ylabel("Total Penjualan (Juta Unit)", fontsize=12)
plt.legend(title="Platform")
plt.grid(True)
plt.show()
```

```python
genre_sales = df.groupby(["Year", "Genre"])[["Global_Sales"]].sum().reset_index()

plt.figure(figsize=(14, 7))
sns.lineplot(data=genre_sales, x="Year", y="Global_Sales", hue="Genre", linewidth=1.5)
plt.title("Tren Penjualan Video Game Berdasarkan Genre", fontsize=14)
plt.xlabel("Tahun Rilis", fontsize=12)
plt.ylabel("Total Penjualan (Juta Unit)", fontsize=12)
plt.legend(title="Genre")
plt.grid(True)
plt.show()
```
